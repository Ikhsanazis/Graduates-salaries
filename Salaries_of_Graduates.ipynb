```python
clean_df= df.dropna()
```

```python
clean_df.isna().sum()
```

```python
clean_df.duplicated().sum()
```

```python
clean_df['Starting Median Salary'].max()
```

```python
#Find College Major with Highest Starting Salaries

clean_df[clean_df['Starting Median Salary'] == clean_df['Starting Median Salary'].max()][["Undergraduate Major","Starting Median Salary"]]
```

```python
#Find College Major with Highest Mid-Career Median Salary

clean_df[clean_df['Mid-Career Median Salary'] == clean_df['Mid-Career Median Salary'].max()][["Undergraduate Major","Mid-Career Median Salary"]]
```

```python
#Find College Major with Lowest Salary

clean_df[clean_df['Starting Median Salary'] == clean_df['Starting Median Salary'].min()][["Undergraduate Major","Starting Median Salary"]]
```

```python
#Find College Major with Lowest Mid-Career Median Salary

clean_df[clean_df['Mid-Career Median Salary'] == clean_df['Mid-Career Median Salary'].min()][["Undergraduate Major","Mid-Career Median Salary"]]
```

```python
#Lowest Risk Majors
#A low-risk major is a degree where there is a small difference between the lowest and highest salaries.

clean_df["Spread"]=clean_df['Mid-Career 90th Percentile Salary']-clean_df['Mid-Career 10th Percentile Salary']
```

```python
low_risk = clean_df.sort_values('Spread')
low_risk.head()
```

```python
#Most Potential
most_potential=clean_df.sort_values('Mid-Career 90th Percentile Salary',ascending=False)
most_potential[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

```python
#Majors with the Greatest Spread in Salaries
highest_spread = clean_df.sort_values('Spread', ascending=False)
highest_spread[['Undergraduate Major', 'Spread']].head()
```
